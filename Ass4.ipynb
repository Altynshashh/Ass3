### Preparation part

```python

# Let's open and read our txt file named ReadingFile.txt
in_text = open('ReadingFile.txt', 'r')
in_text = in_text.read()
in_text

# Now we find the total number of all uniques
len(set(in_text))

# Let's convert our variable to list for using method sort()
w_text = list(set(in_text))
w_text.sort()
w_text

# Now we create a dictionary of uniques and their frequencies
our_dict = {
    symbol: len([a for a in in_text 
                 if a==symbol]) 
    for symbol in w_text}

# Processing
for e in our_dict:
    our_dict[e] = round(our_dict[e]/len(in_text), 3)
our_dict = sorted(our_dict.items(), key=lambda x: x[1])
our_dict = dict(our_dict)

df = {
    symbol: [round(len([a for a in in_text 
                        if a==symbol])/len(in_text), 3), 'None', 'None', ''] 
    for symbol in w_text}
df = sorted(df.items(), key=lambda a: a[1])
df = dict(df)

def Huffman(our_dict):
    global df
    min1 = list(our_dict)[0]
    min2 = list(our_dict)[1]
    
    df[f'node{int((len(df) - len(our_dict))/2)}'] = [our_dict[min1] + our_dict[min2],
                                                            min1, min2, '']
    
    our_dict[f"node{int(((len(df) - 1) - len(our_dict))/2)}"] = our_dict[min1] + our_dict[min2] 
    
    del our_dict[min1]
    del our_dict[min2]
    if(len(our_dict)==1):
        return (our_dict, df) 
    else:
        return Huffman(our_dict)

agash = Huffman(our_dict)[1]

agash = dict(sorted(agash.items(), key=lambda a: a[1], reverse = True))

for key in agash:
    if agash[key][1] != 'None':
        agash[agash[key][1]][-1] = agash[key][-1] + '0'
        agash[agash[key][2]][-1] = agash[key][-1] + '1'
    else:
        pass
```

```python
#1 Let's open and see the encoded text
original = ''
for s in in_text:
    original = original + agash[s][-1]
original
```

```python
#2 #showing all steps with nodes
counter = ''
result = ['Step 1:']
i = 2
transformed = ''
for b in original:
    counter = counter + b
    for key in agash:
        if (agash[key][-1] == counter) & (agash[key][1] == 'None'):
            result.append('Symbol: ' + counter + ' - "' + key + '"')
            result.append(f'Step {i}:')
            transformed = transformed + key
            counter = ''
            i = i + 1
        elif agash[key][-1] == counter:
            result.append(counter + ': "' + key + '": "' + agash[key][1] + '", "' + agash[key][2])

result = result[:-1]
result
```

```python
#3 Making comparison between original and decoded texts
print("Decoded Text: " + transformed,'\n')
print("Original Text: " + in_text,'\n')
print(transformed == in_text)
```
